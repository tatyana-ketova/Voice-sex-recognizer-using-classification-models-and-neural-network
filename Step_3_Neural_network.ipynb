```python
df_train=shuffled.head(1700) #I divide the dataframe into training and test parts
df_test=shuffled.tail(418)
```

```python
data=df_train.iloc[:,:-1].to_numpy()
target=df_train.iloc[:,-1].to_numpy()
data_test=df_test.iloc[:,:-1].to_numpy()
target_test=df_test.iloc[:,-1].to_numpy()
```

```python
dataset= [(data[i][None,...],target[i]) for i in range(len(target))]
print(dataset)
```

```python
INPUT_DIM=32 # number of input neurons
OUT_DIM=2 # number of output neurons
H_DIM=14 #number of neurons in the middle layer
```

```python
def relu(t): #activation function
    return np.maximum(t,0)
```

```python
def softmax(t): #function to get probability
    out=np.exp(t)
    return out/np.sum(out)
```

```python
def softmax_batch(t):#function to get probability
    out=np.exp(t)
    return out/np.sum(out,axis=1, keepdims=True)
```

```python
def sparce_cross_entropy_batch(z,y):
    return np.log(np.array([z[j,y[j]] for j in range(len(y))]))
```

```python
def to_full(y,num_classes):
    y_full=np.zeros((1,num_classes))
    y_full[0,y]=1
    return y_full
```

```python
def to_full_batch(y,num_classes):
    y_full=np.zeros((len(y),num_classes))
    for j,yj in enumerate(y):
        y_full[j,yj]=1
    return y_full  
```

```python
def relu_deriv(t):
    return(t>=0).astype(float)
```

```python
x=np.random.randn(1, INPUT_DIM)
y=random.randint(0, OUT_DIM-1)
```

```python
W1=np.random.randn(INPUT_DIM, H_DIM)# weight matrix
b1=np.random.randn(1, H_DIM)#vector1
W2=np.random.randn(H_DIM, OUT_DIM)# weight vector
b2=np.random.randn(1, OUT_DIM)#vector2
```

```python
W1 = (W1 - 0.5) * 2 * np.sqrt(1/INPUT_DIM)
b1 = (b1 - 0.5) * 2 * np.sqrt(1/INPUT_DIM)
W2 = (W2 - 0.5) * 2 * np.sqrt(1/H_DIM)
b2 = (b2 - 0.5) * 2 * np.sqrt(1/H_DIM)
```

```python
#learning rate
ALPHA=0.001
#number of epochs
NUM_EPOCHS=100
BATCH_SIZE=50
loss_arr=[]
```

```python
for ep in range(NUM_EPOCHS):
    random.shuffle(dataset)
    for i in range(len(dataset)//BATCH_SIZE):
        batch_x,batch_y=zip(*dataset[i*BATCH_SIZE:i*BATCH_SIZE+BATCH_SIZE])
        x=np.concatenate(batch_x,axis=0)
        y=np.array(batch_y)
        
        #Forward
        t1=x@W1+b1
        h1=relu(t1)
        t2=h1@W2+b2
        z=softmax_batch(t2)
        E=np.sum(sparce_cross_entropy_batch(z,y)) 
        
        #Backward
        y_full=to_full_batch(y,OUT_DIM)
        dE_dt2=z-y_full
        dE_dW2=h1.T @ dE_dt2
        dE_db2=np.sum(dE_dt2,axis=0,keepdims=True)
        dE_dh1=dE_dt2 @ W2.T
        dE_dt1=dE_dh1*relu_deriv(t1)
        dE_dW1=x.T @ dE_dt1
        dE_db1=np.sum(dE_dt1,axis=0,keepdims=True)
        
        #Update
        W1=W1-ALPHA*dE_dW1
        b1=b1-ALPHA*dE_db1
        W2=W2-ALPHA*dE_dW2
        b2=b2-ALPHA*dE_db2
        
        loss_arr.append(E)
```

```python
def predict(x):
    t1=x@W1+b1
    h1=relu(t1)
    t2=h1@W2+b2
    z=softmax_batch(t2)
    return z
```

```python
def calc_accuracy():
    correct=0
    for x,y in dataset:
        z=predict(x)
        y_pred=np.argmax(z)
        if y_pred==y:
            correct+=1
    acc=correct/len(dataset)  
    return acc
```

```python
accuracy=calc_accuracy()
print('Accuracy:',accuracy)
```
